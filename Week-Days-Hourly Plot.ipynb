```python
# station_list
slst = pd.read_csv("/content/drive/My Drive/Data Links/stations_corrected.csv")
slst.head()
```

```python
slst_df = pd.DataFrame(slst)
```

# Weekday Count Calculate

```python
weekday_start_date = '02-MAY-16 00:00:00'
weekday_end_date = '06-MAY-16 23:59:59'
```

```python
mask = (mts_df['ENTRY_DTM'] > weekday_start_date) & (mts_df['ENTRY_DTM'] <= weekday_end_date)
mts_df_weekdays_entry = mts_df.loc[mask]
```

```python
mts_df_weekdays_entry.head()
```

```python
mts_df_weekdays_entry.size
```

```python
mask = (mts_df['EXIT_DTM'] > weekday_start_date) & (mts_df['EXIT_DTM'] <= weekday_end_date)
mts_df_weekdays_exit = mts_df.loc[mask]
```

```python
mts_df_weekdays_exit.head()
```

```python
mts_df_weekdays_exit.size
```

```python
# Weekday Entry count
weekday_entry_count = mts_df_weekdays_entry['ENTRY_MSTN_ID'].value_counts().reset_index()
weekday_entry_count.columns = ['STATION_ID', 'ENTRY_COUNT']
weekday_entry_count.head()
```

```python
# Weekday Exit count
weekday_exit_count = mts_df_weekdays_exit['EXIT_MSTN_ID'].value_counts().reset_index()
weekday_exit_count.columns = ['STATION_ID', 'EXIT_COUNT']
weekday_exit_count.head()
```

```python
weekday_df = slst_df
weekday_df = pd.merge(weekday_df, weekday_entry_count, on=['STATION_ID'], how='left')
weekday_df = pd.merge(weekday_df, weekday_exit_count, on=['STATION_ID'], how='left')
weekday_df
```

# Plot Weekdays

```python
fig_weekday_entry_sc = px.scatter_mapbox(weekday_df, lat="lat", lon="lon", color_discrete_sequence=['green'], size="ENTRY_COUNT",
                  color_continuous_scale=px.colors.cyclical.IceFire, size_max=15, zoom=10)
fig_weekday_entry_sc.show()
```

```python
fig_weekday_exit_sc = px.scatter_mapbox(weekday_df, lat="lat", lon="lon", color_discrete_sequence=['red'], size="EXIT_COUNT",
                  color_continuous_scale=px.colors.cyclical.IceFire, size_max=15, zoom=10)
fig_weekday_exit_sc.show()
```
